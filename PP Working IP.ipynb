```python
result = driver.find_elements_by_xpath("//ui-scoreboard-runner[@class='ui-scoreboard-runner ui-scoreboard-runner__home']/span")    
resultaway = driver.find_elements_by_xpath("//ui-scoreboard-runner[@class='ui-scoreboard-runner ui-scoreboard-runner__away']/span")
resultoddshome= driver.find_elements_by_xpath("//ng-transclude[@class='button__content']/span")
my_columns = ['Home Team','Away Team','Home Team Odds','Draw Odds','Away Team Odds']
final_dataframe = pd.DataFrame(columns = my_columns)

string_new = ''
for i in result:
    string_new = string_new + i.text + ","

def Convert(string):
    li = list(string.split(","))
    return li

stringhome = Convert(string_new)
string_new1 = ''
for i in resultaway:
    string_new1 = string_new1 + i.text + ","

stringaway = Convert(string_new1)
string_new2 = ''
for i in resultoddshome:
    string_new2 = string_new2 + i.text + ","

stringodds = Convert(string_new2)
every_third_element = stringodds[::3]
every_second_element = stringodds[1::3]
every_first_element = stringodds[2::3]
i = " "
every_second_element.append(i)
every_first_element.append(i)
final_dataframe['Home Team'] = stringhome
final_dataframe['Away Team'] = stringaway
final_dataframe['Home Team Odds'] = every_third_element
final_dataframe['Draw Odds'] = every_second_element
final_dataframe['Away Team Odds'] = every_first_element
final_dataframe
```
